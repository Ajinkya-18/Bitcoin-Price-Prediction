## Feature Engineering

```python
splitted = df['timestamp'].str.split(' ', expand=True)
df['date'] = splitted[0].astype('str')
df['time'] = splitted[1].astype('str')
```

```python
df = df.drop('timestamp', axis=1)
```

```python
splitted1 = df['date'].str.split('-', expand=True)
df['year'] = splitted1[0].astype('int')
df['month'] = splitted1[1].astype('int')
df['day'] = splitted1[2].astype('int')
```

```python
# saved the date and time column in separate series for future reference before deleting from the dataframe
date = pd.Series(df['date'])
time = pd.Series(df['time'])
```

```python
df = df.drop('date', axis=1)
df = df.drop('time', axis=1)
```

```python
df.head()
```

```python
df['open-close'] = df['open'] - df['close']
df['low-high'] = df['low'] - df['high']
df['target'] = np.where(df['close'].shift(-1) > df['close'], 1, 0) # checks if the price went up (1) or down (0)
```

```python
df['is_quarter_end'] = np.where((df['month']%3)==0,1,0)
```

```python
plt.pie(df['target'].value_counts().values, labels = [0,1], autopct = "%.2f%%")
plt.show()
```

```python
corr_mat = df.corr()

plt.figure(figsize = (15,15))
sns.heatmap(corr_mat, annot = True, cbar=True)
plt.show()
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(corr_mat > 0.9, annot=True, cbar=True)
plt.show()
```

## Model Training

```python
X = df[['open-close', 'low-high', 'is_quarter_end', 'taker_buy_base_asset_volume', 
        'taker_buy_quote_asset_volume']]
Y = df['target']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print(X_train_scaled.shape, X_test_scaled.shape)
```

```python
from pickle import dump
with open('../models/fitted_scaler.pkl', 'wb') as f:
    dump(scaler, f)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
lc = LogisticRegression()
knc = KNeighborsClassifier()
dtc = DecisionTreeClassifier()
hgbc = HistGradientBoostingClassifier()
```

```python
lc.fit(X_train_scaled, Y_train)
knc.fit(X_train_scaled, Y_train)
dtc.fit(X_train_scaled, Y_train)
hgbc.fit(X_train_scaled, Y_train)
```

```python
lc_preds = lc.predict(X_test_scaled)
knc_preds = knc.predict(X_test_scaled)
dtc_preds = dtc.predict(X_test_scaled)
hgbc_preds = hgbc.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_metrics = {}
model_preds = [lc_preds, knc_preds, dtc_preds, hgbc_preds]
model_names = ['Logistic Regression', 'K-Neighbors Classifier', 'Decision Tree Classifier', 
               'HistGradient Boosting Classifier']

accuracy_scores = [accuracy_score(preds, Y_test) for preds in model_preds]

for model, acc in zip(model_names, accuracy_scores):
    accuracy_metrics[model] = round(acc, 4)
```
